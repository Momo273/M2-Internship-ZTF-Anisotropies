```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import scipy.optimize as optimize
from SN_distribution import SkyMap
from SN_distribution import Patching_Map
from Selections import Criteria_selection
from Selections import Good_sampling
from Selections import salt_2_selection
from Hubble import Hubble_fit
from Hubble import Hubble_fit_patch
import iminuit
```

```python
df_lc = pd.read_csv("lc/ZTF2.csv")
```

```python
mask_flux = (df_lc['flux'] > 0)
mask_band_r = (df_lc['band'] == 'ztfr')
mask_band_g = (df_lc['band'] == 'ztfg')
mask_band_i = (df_lc['band'] == 'ztfi')

plt.errorbar(df_lc['time'][mask_flux][mask_band_r], df_lc['flux'][mask_flux][mask_band_r]
             , marker='.', yerr = df_lc['fluxerr'][mask_flux][mask_band_r], label = 'Band r', 
             linestyle="None", capsize = 2, zorder = 1, color = 'red', ecolor = 'salmon', elinewidth = 0.5)
plt.errorbar(df_lc['time'][mask_flux][mask_band_g], df_lc['flux'][mask_flux][mask_band_g], label = 'Band g'
             , marker='.', yerr = df_lc['fluxerr'][mask_flux][mask_band_g], linestyle="None", 
              capsize = 2, zorder = 1, color = 'green', ecolor = 'seagreen', elinewidth = 0.5)
plt.errorbar(df_lc['time'][mask_flux][mask_band_i], df_lc['flux'][mask_flux][mask_band_i], label = 'Band i',
             marker='.', yerr = df_lc['fluxerr'][mask_flux][mask_band_i], linestyle="None", 
              capsize = 2, zorder = 1, color = 'orange',  ecolor = 'goldenrod', elinewidth = 0.5)
plt.xlabel('Time (MJD)')
plt.ylabel('Flux (ADU)')
plt.legend(shadow = True);
```
